```python
''' You can simulate sending the information to the database by using the ModelProcessorOperator Class

Simply start a new instance and  pass the entry as a parameter and invoke the orquestrator method.

ex: ModelProcessorOperator(records).orquestrator()

But, you wont be able to actually upload info in the BigQuery database due to billing and security concerns.

So, pass the parameter upload_to_bq=False, to simulate the loop without actually uploading to the BQ Database.

'''
```
